### RNN with Attention (Possible solution)

```python
# -*- coding: utf-8 -*-
"""
Wind Turbine Power Prediction with RNNs + Attention

This lab uses an Attention-enhanced Recurrent Neural Network to predict wind turbine power output.
"""

# Step 0: Load libraries & Data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout, Layer
from tensorflow.keras.callbacks import EarlyStopping
import math

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# Step 1: Prepare the Data
# Load Dataset
df = pd.read_csv("T1.csv")
df.columns = df.columns.str.strip()
print("Shape:", df.shape)
display(df.head())
```

```python
# Selecting Relevant Features
# Select relevant features
features = ["LV ActivePower (kW)", "Wind Speed (m/s)", "Theoretical_Power_Curve (KWh)"]
data = df[features].copy()

# Normalizing with MinMaxScaler
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)
data_scaled = pd.DataFrame(data_scaled, columns=features)

print("\nScaled feature ranges (should be ~0 to 1):")
display(data_scaled.agg(["min", "max"]))
```

```python
# Step 2: Create sequences
SEQ_LEN = 24
values = data_scaled.values
target_col = 0

X, y = [], []
for i in range(SEQ_LEN, len(values)):
    X.append(values[i-SEQ_LEN:i, :])
    y.append(values[i, target_col])

X = np.array(X)
y = np.array(y)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
# Step 3: Split the data

train_size = int(0.8 * len(X))

X_train = X[:train_size]
X_test  = X[train_size:]

y_train = y[:train_size]
y_test  = y[train_size:]

# Check shapes
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
# Step 4: Define a Custom Attention Layer


class Attention(Layer):
    """
    Simple additive attention over time steps.
    Input:  sequences of hidden states from RNN/LSTM, shape (batch, timesteps, hidden_dim)
    Output: context vector (batch, hidden_dim) and attention weights (batch, timesteps)
    """
    def __init__(self, **kwargs):
        super().__init__(**kwargs)

    def build(self, input_shape):
        # input_shape = (batch, timesteps, hidden_dim)
        hidden_dim = input_shape[-1]

        # Trainable parameters to compute attention scores
        self.W = self.add_weight(
            name="W",
            shape=(hidden_dim, hidden_dim),
            initializer="glorot_uniform",
            trainable=True
        )
        self.b = self.add_weight(
            name="b",
            shape=(hidden_dim,),
            initializer="zeros",
            trainable=True
        )
        self.v = self.add_weight(
            name="v",
            shape=(hidden_dim, 1),
            initializer="glorot_uniform",
            trainable=True
        )
        super().build(input_shape)

    def call(self, inputs):
        # inputs: (batch, timesteps, hidden_dim)

        # Score each time step
        # score: (batch, timesteps, hidden_dim)
        score = tf.tanh(tf.tensordot(inputs, self.W, axes=[[2],[0]]) + self.b)

        # attention_logits: (batch, timesteps, 1)
        attention_logits = tf.tensordot(score, self.v, axes=[[2],[0]])

        # attention_weights: (batch, timesteps, 1)
        attention_weights = tf.nn.softmax(attention_logits, axis=1)

        # context_vector: weighted sum of inputs across timesteps
        # (batch, timesteps, hidden_dim) * (batch, timesteps, 1) -> (batch, timesteps, hidden_dim)
        # sum over timesteps -> (batch, hidden_dim)
        context_vector = tf.reduce_sum(attention_weights * inputs, axis=1)

        # return weights squeezed to (batch, timesteps) for easier plotting
        return context_vector, tf.squeeze(attention_weights, axis=-1)
```

```python
# Step 5: Build the RNN Model with Attention
timesteps = X_train.shape[1]   # 24
n_features = X_train.shape[2]  # 3

# Input layer
inputs = Input(shape=(timesteps, n_features), name="input_sequence")

# LSTM layers (return_sequences=True so attention can see the all time steps)
x = LSTM(64, return_sequences=True, name="lstm_1")(inputs)
x = Dropout(0.2, name="dropout_1")(x)

# Second LSTM layer (still return_sequences=True for attention)
x = LSTM(32, return_sequences=True, name="lstm_2")(x)
x = Dropout(0.2, name="dropout_2")(x)

# Attention layer -> context vector + attention weights
context_vector, attention_weights = Attention(name="attention")(x)

# Dense layers -> final prediction
x = Dense(32, activation="relu", name="dense_1")(context_vector)
x = Dropout(0.2, name="dropout_3")(x)

outputs = Dense(1, activation="linear", name="prediction")(x)

# Build model
model = Model(inputs=inputs, outputs=outputs, name="LSTM_with_Attention")

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss="mse")

model.summary()
```

```python
# Step 6: Compile and Train the Model

# Early Stopping
early_stop = EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True
)

history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=64,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

plt.figure()
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history['val_loss'], label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.title("Training VS Validation Loss")
plt.legend()
plt.show()
```

```python
# Step 7: Evaluate and Visualize Results

# Predicting on Test Set
y_pred = model.predict(X_test)
# Reshaping y_test because both y_test and y_pred are shape (n, 1)
y_test_2d = y_test.reshape(-1, 1)

# Inverse-transform predictions and true values back to original scale
# MinMaxScaler was fit on 3 columns: [ActivePower, WindSpeed, TheoreticalPower]
# To inverse_transform only ActivePower, we create placeholder arrays with 3 columns.
n_test = len(y_test_2d)

pred_placeholder = np.zeros((n_test, len(features)))
true_placeholder = np.zeros((n_test, len(features)))

# Put predictions and true values into the first column (ActivePower)
pred_placeholder[:, 0] = y_pred.flatten()
true_placeholder[:, 0] = y_test_2d.flatten()

# Inverse transform
pred_inv = scaler.inverse_transform(pred_placeholder)[:, 0]
true_inv = scaler.inverse_transform(true_placeholder)[:, 0]

# Metrics
mse = mean_squared_error(true_inv, pred_inv)
rmse = math.sqrt(mse)
r2 = r2_score(true_inv, pred_inv)

print(f"MSE:  {mse:,.3f}")
print(f"RMSE: {rmse:,.3f}")
print(f"R^2:  {r2:,.4f}")
```
